Que 4.  What are evaluation curves in Detectron2, and how are metrics like mAP and IoU interpreted?

Ans - In Detectron2, evaluation curves and metrics are used to measure how well an object detection or segmentation model performs. The two most important metrics are mAP (mean Average Precision) and IoU (Intersection over Union).

---

1️⃣ IoU (Intersection over Union)

IoU measures how well the predicted bounding box overlaps with the ground-truth bounding box.

It is calculated as:

WhatsApp Image 2026-03-13 at 4.12.55 PM.jpeg

Interpretation:
- IoU = 1.0 → Perfect overlap between prediction and ground truth
- IoU > 0.5 → Usually considered a correct detection
- IoU < 0.5 → Detection considered incorrect

Example

- Ground truth box: object location in dataset
- Predicted box: model prediction

If the overlap area is 70%, then:

IoU = 0.7 → good prediction

In Detectron2, evaluation is often done at multiple IoU thresholds:
- AP50 → IoU ≥ 0.50
- AP75 → IoU ≥ 0.75 (stricter)

---

2️⃣ mAP (Mean Average Precision)

mAP is the main metric used to evaluate object detection models.

Step-by-step idea
1. Model predicts bounding boxes with confidence scores
2. Predictions are sorted by confidence
3. Precision and Recall are calculated
4. A Precision-Recall curve is plotted
5. Average Precision (AP) = area under that curve
6. mAP = mean of AP across all object classes

---

3️⃣ Evaluation Curves in Detectron2

Precision–Recall Curve

This is the main evaluation curve.
- X-axis: Recall
- Y-axis: Precision

Interpretation:
- Curve closer to top-right → better model
- Large area under curve → higher AP

---

Confidence Threshold Curve

Sometimes evaluation also checks performance at different confidence thresholds.

Example:

| Confidence | Result                                        |
| ---------- | --------------------------------------------- |
| 0.9        | very precise but may miss objects             |
| 0.5        | balanced                                      |
| 0.2        | detects more objects but many false positives |


---

4️⃣ Detectron2 COCO Evaluation Metrics

When evaluating using COCO Dataset, Detectron2 prints metrics like:

| Metric   | Meaning                               |
| -------- | ------------------------------------- |
| **AP**   | Average precision across IoU 0.5–0.95 |
| **AP50** | AP at IoU = 0.50                      |
| **AP75** | AP at IoU = 0.75                      |
| **APS**  | AP for small objects                  |
| **APM**  | AP for medium objects                 |
| **APL**  | AP for large objects                  |


Example output:

    bbox AP = 0.41
    AP50 = 0.63
    AP75 = 0.45


Interpretation:
- Model detects objects correctly ~41% on average
- 63% accuracy at IoU 0.5 threshold

---

5️⃣ Simple Interpretation

- IoU → How accurate the predicted box location is
- Precision → How many detected objects are correct
- Recall → How many real objects the model found
- AP → Area under Precision-Recall curve
- mAP → Average AP across all classes

---

✅ In short:

- IoU → box overlap quality
- Precision-Recall curve → detection performance tradeoff
- AP → area under that curve
- mAP → final detection accuracy score

4️⃣ Convert Annotations to COCO Format

Detectron2 requires COCO JSON format.

Install converter:



```python
pip install pascal-voc-writer
```

Or use a Python conversion script.

5️⃣ Python Script to Convert VOC → COCO

```python
import os
import json
import xml.etree.ElementTree as ET

xml_dir = "annotations_xml"
image_dir = "images"

categories = [{"id":1,"name":"car"}]

images = []
annotations = []

ann_id = 1
img_id = 1

for xml_file in os.listdir(xml_dir):

    tree = ET.parse(os.path.join(xml_dir, xml_file))
    root = tree.getroot()

    filename = root.find("filename").text
    width = int(root.find("size/width").text)
    height = int(root.find("size/height").text)

    images.append({
        "id": img_id,
        "file_name": filename,
        "width": width,
        "height": height
    })

    for obj in root.findall("object"):

        xmin = int(obj.find("bndbox/xmin").text)
        ymin = int(obj.find("bndbox/ymin").text)
        xmax = int(obj.find("bndbox/xmax").text)
        ymax = int(obj.find("bndbox/ymax").text)

        w = xmax - xmin
        h = ymax - ymin

        annotations.append({
            "id": ann_id,
            "image_id": img_id,
            "category_id": 1,
            "bbox": [xmin, ymin, w, h],
            "area": w*h,
            "iscrowd": 0
        })

        ann_id += 1

    img_id += 1


coco_format = {
    "images": images,
    "annotations": annotations,
    "categories": categories
}

with open("annotations_coco.json","w") as f:
    json.dump(coco_format,f)
```

6️⃣ Complete Script (All Steps Together)

```python
import os
from detectron2.engine import DefaultTrainer
from detectron2.config import get_cfg
from detectron2 import model_zoo
from detectron2.data.datasets import register_coco_instances

# Register dataset
register_coco_instances(
    "my_dataset",
    {},
    "dataset/annotations/annotations_coco.json",
    "dataset/images"
)

# Create config
cfg = get_cfg()
cfg.merge_from_file(
    model_zoo.get_config_file(
        "COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"
    )
)

cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url(
    "COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"
)

cfg.DATASETS.TRAIN = ("my_dataset",)
cfg.DATASETS.TEST = ()

cfg.DATALOADER.NUM_WORKERS = 2
cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.00025
cfg.SOLVER.MAX_ITER = 3000

cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1

cfg.OUTPUT_DIR = "./output"
os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)

# Train model
trainer = DefaultTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()
```

8️⃣ Complete Inference Script

```python
import cv2
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2 import model_zoo
from detectron2.data import MetadataCatalog

cfg = get_cfg()

cfg.merge_from_file(
    model_zoo.get_config_file(
        "COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"
    )
)

cfg.MODEL.WEIGHTS = "output/model_final.pth"
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
cfg.MODEL.DEVICE = "cpu"

predictor = DefaultPredictor(cfg)

image = cv2.imread("test_image.jpg")

outputs = predictor(image)

v = Visualizer(
    image[:, :, ::-1],
    MetadataCatalog.get(cfg.DATASETS.TRAIN[0]),
    scale=1.2
)

out = v.draw_instance_predictions(outputs["instances"].to("cpu"))

cv2.imshow("Prediction", out.get_image()[:, :, ::-1])
cv2.waitKey(0)
```

✅ Output -
- The image will appear with bounding boxes and labels predicted by the trained Detectron2 model.


Que 10. You are assigned to build a wildlife monitoring system to detect and track different animal species in a forest using Detectron2. Describe the end-to-end pipeline from data collection to deploying the model, and how you would handle challenges like occlusion or nighttime detection.

Ans - To build a wildlife monitoring system that detects and tracks animals using Detectron2, the solution should follow an end-to-end computer vision pipeline from data collection to deployment.

---

1️⃣ Data Collection

Collect images and videos of animals from different sources.

Sources
- Forest camera traps
- Drones
- Public datasets like COCO Dataset or wildlife datasets (Snapshot Serengeti etc.)

Important considerations
- Different species
- Various lighting conditions (day/night)
- Different camera angles
- Multiple animals in frame
- Seasonal variations

---

2️⃣ Data Annotation

Annotate animals in the collected images.

Tools
- LabelImg
- CVAT

Annotation format
- Bounding boxes for each animal
- Class labels (e.g., deer, tiger, elephant)

Convert annotations to COCO Dataset Format because Detectron2 supports it directly.

Example classes:
    
    1  Deer
    2  Tiger
    3  Elephant
    4  Monkey

3️⃣ Data Preprocessing

Before training the model:

Cleaning
- Remove blurry or corrupted images

Dataset split
- Training: 70%
- Validation: 20%
- Testing: 10%

Data augmentation
- Horizontal flip
- Rotation
- Random cropping
- Brightness/contrast change

This helps the model generalize to different forest environments.

---

4️⃣ Model Selection

Choose a suitable architecture from Detectron2 model zoo.

Common choices:
- Faster R-CNN for high accuracy
- RetinaNet for balanced performance
- Mask R-CNN if segmentation is required

Example model:

    Faster R-CNN R50-FPN

Load pretrained weights trained on COCO to apply transfer learning.

---

5️⃣ Model Training

Train the model on the annotated dataset.

Key configuration parameters:

    Learning rate
    Batch size
    Number of iterations
    Number of classes
    Score threshold

Training steps:
1. Register dataset
2. Load pretrained weights
3. Train using Detectron2 Trainer
4. Save trained weights

---

6️⃣ Model Evaluation

Evaluate model performance using metrics such as:
- mAP (Mean Average Precision)
- Precision
- Recall
- IoU (Intersection over Union)

Evaluation is done on the validation/test dataset.

Example goal:

    mAP > 0.75

---

7️⃣ Animal Tracking

Detection identifies animals in each frame.

Tracking links the same animal across frames.

Common tracking algorithms:
- SORT
- DeepSORT

Output example:

    Tiger ID 1 → moving east
    Elephant ID 2 → moving north

This helps analyze movement patterns and population behavior.

---

8️⃣ Handling Key Challenges

Occlusion (animals hidden behind trees)

Solutions:
- Use multi-frame tracking
- Use larger datasets with partial visibility
- Use instance segmentation models like Mask R-CNN
- Train with occlusion augmentation

---

Nighttime Detection

Forest cameras often capture images at night.

Solutions:
- Use infrared camera images
- Train with night images
- Apply image enhancement or low-light preprocessing
- Use thermal camera data

---

Small or Distant Animals

Solutions:
- Use Feature Pyramid Networks (FPN)
- Train with higher-resolution images
- Apply multi-scale training

---

9️⃣ Model Optimization

Before deployment:
- Reduce model size
- Optimize inference speed
- Convert model to ONNX or TensorRT

Possible export formats:
- ONNX
- TorchScript

---

🔟 Deployment

Deploy the model for real-time wildlife monitoring.

Possible deployment environments:

Edge Devices
- NVIDIA Jetson
- Raspberry Pi with GPU accelerator

Cloud Systems
- Wildlife monitoring dashboard
- Alert system for endangered species detection

Example workflow:

    Camera Trap → Edge Device → Detectron2 Model → Animal Detection → Cloud Dashboard

Final system capabilities:
